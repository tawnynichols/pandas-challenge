```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Create a pandas series for each grade. 

# Group each series by school
mean_by_grade = school_data_complete.groupby(["school_name", "grade"]).mean()
mean_by_grade.head()
```

```python
# Combine the series into a dataframe for math
pivot_math_df = mean_by_grade.pivot_table(index='school_name', columns='grade', values='math_score')

# Optional: give the displayed data cleaner formatting
pivot_math_df["9th"] = pivot_math_df["9th"].map("{:.2f}".format)
pivot_math_df["10th"] = pivot_math_df["10th"].map("{:.2f}".format)
pivot_math_df["11th"] = pivot_math_df["11th"].map("{:.2f}".format)
pivot_math_df["12th"] = pivot_math_df["12th"].map("{:.2f}".format)
pivot_math_df
```

```python
# Combine the series into a dataframe for reading
pivot_reading_df = mean_by_grade.pivot_table(index='school_name', columns='grade', values='reading_score')

# Optional: give the displayed data cleaner formatting
pivot_reading_df["9th"] = pivot_reading_df["9th"].map("{:.2f}".format)
pivot_reading_df["10th"] = pivot_reading_df["10th"].map("{:.2f}".format)
pivot_reading_df["11th"] = pivot_reading_df["11th"].map("{:.2f}".format)
pivot_reading_df["12th"] = pivot_reading_df["12th"].map("{:.2f}".format)
pivot_reading_df
```
